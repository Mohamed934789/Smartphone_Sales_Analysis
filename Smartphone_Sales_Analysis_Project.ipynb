## <span style="color: #117A65;">üîç Introduction</span>

This project aims to conduct a comprehensive data analysis on a dataset of **2,000 smartphones**, each with detailed attributes such as brand, specifications (RAM, Storage, Screen Size, Battery), pricing, user ratings, and sales performance (Revenue, Profit, Quantity Sold).

The main goal is to extract **key business insights** that help understand performance trends, identify high-performing models or brands, and compare operating systems (Android vs iOS) in terms of profitability and user satisfaction.

- **Matplotlib & Seaborn**: For creating beautiful and insightful visualizations.
- **NumPy**: For numerical computations.
- **Jupyter Notebook**: For interactive data exploration and presentation.
- **Markdown + HTML**: For formatting and styling the report.

---

## <span style="color: #2980B9;">üì¶ Data Preprocessing Steps</span>

Before diving into analysis, data cleaning and preparation were performed to ensure accuracy and consistency:

- ‚úÖ **Missing Values**:
  - Missing `Profit` values were filled as **35% of the Sales_Revenue**.
  - Rows with missing critical data like `Price_USD` or `Sales_Revenue` were **removed** to avoid skewed results.
- ‚úÖ **Consistency Checks**:
  - Checked for unrealistic values (e.g., negative prices or sales).
  - Verified no duplicate `Phone_ID`s.

This preprocessing ensured that the dataset is **clean, reliable, and ready for analysis**.

---

## <span style="color: #117A65;">üéØ Key Performance Indicators (KPIs)</span>

To measure smartphone performance, the following KPIs were calculated:

1. **Total Sales Revenue** ‚Äì Sum of all revenue.
2. **Total Profit** ‚Äì Total profit across all phones.
3. **Number of Units Sold per Brand** ‚Äì Quantity sold by brand.
4. **Average Price per Brand** ‚Äì Mean phone price for each brand.
5. **Top-Rated Phone** ‚Äì Device with the highest rating.
6. **Average Rating per OS** ‚Äì Comparison between Android and iOS.
7. **Most Profitable Phones** ‚Äì Phones generating the highest profit.
8. **Average Profit per Unit Sold** ‚Äì Insight into profit per phone sold.
9. **Android vs iOS Comparison** ‚Äì Sales, revenue, and profit comparison between OSs.

Each KPI was supported by **visualizations** and **statistical insights** to help stakeholders interpret the data more clearly.

---

  <h3 style="color:white;">üîß Step 1: Import Libraries & Load Dataset</h3>
</div>


```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
```

```python
df=pd.read_csv(r'C:\Users\IT SHOP\Downloads\smartphone_sales_dataset.csv')
```

```python
df.head()
```

<div style="background-color:#34495e;padding:10px;border-radius:10px">
  <h3 style="color:white;">üßπ Step 2: Data Cleaning & Handling Missing Values</h3>
</div>


```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['Price_USD'])
```

```python
df["Battery_mAh"].fillna(df["Battery_mAh"].mean(), inplace=True)
```

```python
df["Profit"].fillna(df["Sales_Revenue"] * 0.35,inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Sales_Revenue"] = df["Sales_Revenue"].fillna(df["Price_USD"] * df["Quantity_Sold"])
```

```python
df.isnull().sum()
```

```python
df["Profit"] = df["Profit"].fillna(df["Sales_Revenue"] * 0.35)
```

```python
df.isnull().sum()
```

<div style="background-color:#1abc9c;padding:10px;border-radius:10px">
                        <h3 style="color:white;">                   ‚úÖ Data is Now Clean & Ready for Analysis              </h3>
</div>


<div style="background-color:#3498db;padding:10px;border-radius:10px">
  <h3 style="color:white;">üí∞ Total Sales Revenue and Total Profit</h3>
</div>


```python
total_revenue = df["Sales_Revenue"].sum()
total_profit = df["Profit"].sum()
```

```python
print(f"üîπ Total Sales Revenue: {total_revenue}")
print(f"üîπ Total Profit: {total_profit}")
```

```python
labels = ['Total Sales Revenue', 'Total Profit']
values = [total_revenue, total_profit]

plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=["#4CAF50", "#2196F3"])
plt.title('Sales Revenue vs Profit')
plt.show()
```

<div style="background-color:#9b59b6;padding:10px;border-radius:10px">
  <h3 style="color:white;">üì¶ Units Sold per Brand</h3>
</div>


```python
units_sold_per_brand = df.groupby("Brand")["Quantity_Sold"].sum()
```

```python
units_sold_per_brand
```

```python
plt.figure(figsize=(10, 6))
units_sold_per_brand.plot(kind='bar', color='skyblue')
plt.title('Number of Units Sold per Brand')
plt.xlabel('Brand')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

<div style="background-color:#e74c3c;padding:10px;border-radius:10px">
  <h3 style="color:white;">‚≠ê Top-Rated Phone</h3>
</div>


```python
top_rated_phone = df[df["Rating"] == df["Rating"].max()]
top_rated_phone[['Phone_ID', 'Brand','Rating']]
```

<div style="background-color:#16a085;padding:15px;border-radius:10px;text-align:center">
  <h2 style="color:white;font-size:32px;margin:0;">1. Which brand sold the most phones?</h2>
</div>


```python
df.groupby('Brand')['Quantity_Sold'].sum().sort_values(ascending=False)
```

<div style="background-color:#27ae60;padding:15px;border-radius:10px;text-align:center">
  <h2 style="color:white;font-size:32px;margin:0;">2. Which brand made the most profit?</h2>
</div>


```python
most_profitable_phones = df.sort_values(by="Profit", ascending=False)
most_profitable_phones[['Phone_ID', 'Brand', 'Profit']].head()
```

```python
plt.figure(figsize=(10, 6))
most_profitable_phones.head(10).plot(kind='bar', x='Brand', y='Profit', color='purple')
plt.title('Top 10 Most Profitable Phones')
plt.xlabel('Phone Model')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<div style="background-color:#f39c12;padding:15px;border-radius:10px;text-align:center">
  <h2 style="color:white;font-size:32px;margin:0;">3. Is there a relationship between RAM and sales?</h2>
</div>


```python
df[['RAM_GB', 'Quantity_Sold']].corr()
```

```python
sns.scatterplot(data=df, x='RAM_GB', y='Quantity_Sold')
```

<div style="background-color:#e67e22;padding:15px;border-radius:10px;text-align:center">
  <h2 style="color:white;font-size:32px;margin:0;">4. Does battery size affect sales?</h2>
</div>


```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Battery_mAh'], df['Quantity_Sold'], color='teal')
plt.title('Battery Capacity vs Units Sold')
plt.xlabel('Battery Capacity (mAh)')
plt.ylabel('Quantity Sold')
plt.show()
```

<div style="background-color:#2980b9;padding:15px;border-radius:10px;text-align:center">
  <h2 style="color:white;font-size:32px;margin:0;">5. Which OS has the higher average rating ‚Äì Android or iOS?</h2>
</div>


```python
avg_rating_per_os = df.groupby("OS")["Rating"].mean()
avg_rating_per_os
```

```python
plt.figure(figsize=(6, 6))
avg_rating_per_os.plot(kind='bar', color=['green', 'orange'])
plt.title('Average Rating per OS')
plt.xlabel('Operating System')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

<div style="background-color:#9b59b6;padding:15px;border-radius:10px;text-align:center">
  <h2 style="color:white;font-size:32px;margin:0;">6. Do higher-priced phones sell more?</h2>
</div>


```python
plt.figure(figsize=(8,8))
plt.scatter(x=df['Price_USD'],y=df['Quantity_Sold'],color='teal')
plt.ylabel('Quantity_Sold')
plt.xlabel('Price_USD')
plt.title('Price_USD vs Units Sold')
plt.show()
```

<div style="background-color:#34495e;padding:10px;border-radius:10px">
  <h3 style="color:white;">üìä Average Profit per Unit Sold</h3>
</div>


```python
df['Avg_Profit_per_Unit'] = df['Profit'] / df['Quantity_Sold']
avg_profit_per_unit = df['Avg_Profit_per_Unit'].mean()
avg_profit_per_unit
```

<div style="background-color:#8e44ad;padding:10px;border-radius:10px">
  <h3 style="color:white;">ü§ñüì± Android vs iOS: Revenue, Profit & Sales</h3>
</div>


```python
os_comparison = df.groupby("OS")[["Sales_Revenue", "Profit", "Quantity_Sold"]].sum()
os_comparison
```

```python
os_comparison.plot(kind='bar', figsize=(10, 6), color=['#FF5722', '#4CAF50'])
plt.title('Android vs iOS: Revenue, Profit, Sales')
plt.xlabel('Operating System')
plt.ylabel('Amount')
plt.xticks(rotation=0)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Rating'], df['Quantity_Sold'], color='purple')
plt.title('Rating vs Quantity Sold')
plt.xlabel('Rating')
plt.ylabel('Quantity Sold')
plt.tight_layout()
plt.show()
```
